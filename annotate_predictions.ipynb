```python
patterns = {}
patterns['add_sentence'] = r'(\[[\w\s]+\.\])'
patterns['subtract_sentence'] = r'(\{[\w\s]+\.\})'
patterns['add_connected_sentences'] = r'\(\[.*?\]\)'
patterns['subtract_connected_sentences'] = r'\(\{.*?\}\)'
patterns['replace_phrase'] = r'\<.*?\>'
patterns['all_adds'] = r'\[.*?\]'
patterns['all_subtracts'] = r'\{.*?\}'
patterns['all_edits'] = r'[\[\{][\w\s\.]*[\]\}]'


def find_single_adds(patterns, annotation, task_id):
    all_adds = re.findall(patterns['all_adds'], annotation)
    all_adds = [sent.strip('[]') for sent in all_adds]
    
    connected_adds = re.findall(patterns['add_connected_sentences'], annotation)
    connected_adds = [[sent.strip('[]') for sent in text.strip("()").split('] [')] for text in connected_adds]
    connected_adds = [sent for text in connected_adds for sent in text]
    
    replaces = re.findall(patterns['replace_phrase'], annotation)
    try:
        replaces = [text.strip("<>").split('},[')[1].strip(']') for text in replaces]
    except:
        import pdb; pdb.set_trace()
    return list(set(all_adds) - set(connected_adds) - set(replaces))

def find_single_subtracts(patterns, annotation, task_id):
    all_subtracts = re.findall(patterns['all_subtracts'], annotation)
    all_subtracts = [sent.strip('{}') for sent in all_subtracts]
    
    connected_subtracts = re.findall(patterns['subtract_connected_sentences'], annotation)
    connected_subtracts = [[sent.strip('{}') for sent in text.strip("()").split('} {')] for text in connected_subtracts]
    connected_subtracts = [sent for text in connected_subtracts for sent in text]
    
    replaces = re.findall(patterns['replace_phrase'], annotation)
    try:
        replaces = [text.strip("<>").split('},[')[0].strip('{') for text in replaces]
    except:
        import pdb; pdb.set_trace()
    
    return list(set(all_subtracts) - set(connected_subtracts) - set(replaces))

def find_segments(annotation):
    res = [segment.strip() for segment in re.split('\s\.|\s\,',annotation)]
    res.remove('')
    return res


# number of segments
# number of high-level subgoals
# number of low-level actions

# NORMALIZE
# dist - num segments
# dist - sent lengths
# dist - num replaces
# dist - num adds
# dist - num subtracts
# dist - num isolated adds
# dist - num isolated subtracts
# dist - len of add chunks
# dist - len of subtract chunks 
# dist - num of replace chunks
# dist - number of noun replaces
```
