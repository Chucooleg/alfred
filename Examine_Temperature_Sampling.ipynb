## Aggregate Statistics

```python
def compare_stats(metrics):
    i = 0
    
    for mtr, t in metrics:
        print('\n----TEMPERATURE = {}----------------------------------------------'.format(t))
        print('Sampling Perplexity Mean = {}'.format(round(np.nanmean(mtr[i]['sampling_perplexity']),3)))
        print('Sampling Perplexity Median = {}'.format(round(np.nanmedian(mtr[i]['sampling_perplexity']),3)))
        print('Sampling Perplexity Max = {}'.format(round(mtr[i]['sampling_perplexity_max'],3)))
        print('Sampling Perplexity Min = {}'.format(round(mtr[i]['sampling_perplexity_min'],3)))
        print()
        print('BLEU Mean = {}'.format(round(mtr[i]['BLEU_mean'],3)))
        print('BLEU Median = {}'.format(round(mtr[i]['BLEU_median'],3)))
        print('BLEU Max = {}'.format(round(mtr[i]['BLEU_max'],3)))
        print('BLEU Min = {}'.format(round(mtr[i]['BLEU_min'],3)))
        print()
        print('Sentence Length Mean = {}'.format(round(mtr[i]['sentence_length_mean'],3)))
        print('Sentence Length Median = {}'.format(round(mtr[i]['sentence_length_median'],3)))
        print('Sentence Length Max = {}'.format(round(mtr[i]['sentence_length_max'],3)))
        print('Sentence Length Min = {}'.format(round(mtr[i]['sentence_length_min'],3)))
        print()
        print('Unique word types = {}'.format(len(set(mtr[i]['word_counts']))))
```

## Examples

```python
def perplexity(probs):
    return 2**(-1*np.sum(np.log2(probs))/len(probs))

def compare(preds, k):
    num_subgoal = len(preds[0][0][k]['p_lang_probs'][0])
        
    for i in range(num_subgoal):
        print ('\nSubgoal {}'.format(i))

        for p, t in preds:

            print('\n----TEMPERATURE = {}----------------------------------------------'.format(t))

            sorted_predictions = sorted([
                (
                    perplexity(p[k]['p_lang_probs'][j][str(i)]),
                    p[k]['p_lang_instr'][j][str(i)],
                )
                for j in range(len(p[k]['p_lang_probs']))
            ], key=lambda x: x[0], reverse=False)


            for perp, instr in sorted_predictions:
                print (round(perp,3), '\n', instr)
```
