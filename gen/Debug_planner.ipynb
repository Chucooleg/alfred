## Examine diversity in generated trajectories

```python
def grid_display(list_of_images, list_of_titles=[], no_of_columns=2, figsize=(10,10)):

    fig = plt.figure(figsize=figsize)
    column = 0
    for i in range(len(list_of_images)):
        column += 1
        #  check for end of column and create a new figure
        if column == no_of_columns+1:
            fig = plt.figure(figsize=figsize)
            column = 1
        fig.add_subplot(1, no_of_columns, column)
        plt.imshow(list_of_images[i])
        plt.axis('off')
        if len(list_of_titles) >= len(list_of_images):
            plt.title(list_of_titles[i])

def compare_traj_outputs(traj_outputs):
    traj_data_all = []
    
    for traj_output in traj_outputs:
        with open(os.path.join(traj_output, 'traj_data.json'), 'r') as f:
            traj_data_all.append(json.load(f))
            
    # list task params
    print('---------------------------')
    print(traj_data_all[0]['pddl_params'])
    print('---------------------------')
    for traj_data in traj_data_all:
        high_pddl = traj_data['plan']['high_pddl']
        for step in high_pddl:
            print(step['discrete_action'])
        print('')
    print('---------------------------')        
#     for traj_data in traj_data_all:
#         low_actions = traj_data['plan']['low_actions']
#         for step in low_actions:
#             print(step['discrete_action']['action'])
#         print('')
#     print('---------------------------')       
#     for traj_output in traj_outputs:
#         print(os.path.join(traj_output, 'video.mp4'))
#         # print images

def get_image_names(traj_data, end_of_subgoals_only=False):
    idx_type = 'high_idx' if end_of_subgoals_only else 'low_idx'
    curr_idx = 0
    image_names = [traj_data['images'][0]['image_name']]
    for image_dict in traj_data['images']:
        if image_dict[idx_type] > curr_idx:
            image_names.append(image_dict['image_name'])
            curr_idx = image_dict[idx_type]
    image_names.append(traj_data['images'][-1]['image_name'])
    return image_names

def compare_image_outputs(traj_outputs, end_of_subgoals_only=False):
    traj_data_all = []
    
    for traj_output in traj_outputs:
        with open(os.path.join(traj_output, 'traj_data.json'), 'r') as f:
            traj_data_all.append(json.load(f))
            
    print('---------------------------')
    for traj_data, traj_output_path in zip(traj_data_all, traj_outputs):
        image_names = get_image_names(traj_data, end_of_subgoals_only)
        image_paths = [os.path.join(traj_output_path, 'raw_images',image_name) for image_name in image_names]
        images = [cv2.imread(image_path) for image_path in image_paths]
        images = [cv2.cvtColor(image, cv2.COLOR_BGR2RGB) for image in images]   
        grid_display(images, [], 5, (10,10))
        
```

```python
compare_traj_outputs(traj_outputs)
```

```python
compare_image_outputs(traj_outputs, end_of_subgoals_only=True)
```
